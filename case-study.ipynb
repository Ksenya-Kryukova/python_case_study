Как критики относятся к спортивным играм? 

```python
import pandas as pd
```

```python
df = pd.read_csv('video_games_sales.csv')
df
```

H0: средняя удовлетворенность критиков спортивными видеоиграм >= 5.0
H1: средняя удовлетворенность критиков спортивными видеоиграм < 5.0

```python
from scipy import stats as st
```

```python
import numpy as np
```

```python
#выделяем только те игры, у которых жанр - спорт + где есть оценка критиков
sport = df.loc[df['Genre'] == 'Sports']
sports = sport.loc[sport['Critic_Score'] >= 0]
sports
```

```python
alpha = 0.05
result = st.ttest_1samp(sports['Critic_Score'], 5.0)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, критикам не нравятся спортивные видеоигры')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критикам нравятся спортивные видеоигры')
```

```python
sport.Critic_Score.mean()
```

Средняя оценка спортивных видеоигр = 7.3, также t-тест показал, что p-value > alpha. Следовательно, критикам нравятся видеоигры

Критикам нравятся больше игры на PC или на PS4?

H0: средняя удовлетворенность критиков PC и PS4 одинаковая
H1: средняя удовлетворенность критиков PC и PS4 различается

```python
PC = df.loc[df['Platform'] == 'PC'].loc[df['Critic_Score'] >= 0]
PS4 = df.loc[df['Platform'] == 'PS4'].loc[df['Critic_Score'] >= 0]
```

```python
result = st.ttest_ind(PC['Critic_Score'], PS4['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по-разному относятся к платформам PC и PS4')
else:
    print('Не отвергаем нулевую гипотезу, критики одинаково относятся к платформам PC и PS4')
```

t-тест показал, что критики по-разному относятся к видеогиграм на платформах PC и PS4, более того, поскольку t-критерий отрицательный, то можно сделать вывод, что критики выше оценивают видеоигры на платформе PS4

Критикам больше нравятся стрелялки или стратегии?

```python
df.groupby('Genre')['Critic_Score'].mean()
```

```python
H0: средняя удовлетворенность критиков стрелялками и стратегиями одинаковая
H1: средняя удовлетворенность критиков стрелялками и стратегиями различается
```

```python
Shooter = df.loc[df['Genre'] == 'Shooter'].loc[df['Critic_Score'] >= 0]
Strategy = df.loc[df['Genre'] == 'Strategy'].loc[df['Critic_Score'] >= 0]
```

```python
result = st.ttest_ind(Shooter['Critic_Score'], Strategy['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по-разному относятся к стрелялкам и стратегиям')
else:
    print('Не отвергаем нулевую гипотезу, критики одинаково относятся к стрелялкам и стратегиям')
```

t-тест показал, что критики одинаково (вернее, что нет статистической значимости) относятся к стрелкам и стратегиям, более того, но поскольку t-критерий отрицательный, то можно сделать вывод, что критики немного выше оценивают стратегии

```python
Shooter.Critic_Score.mean() < Strategy.Critic_Score.mean()
```

```python
Shooter.Critic_Score.mean() - Strategy.Critic_Score.mean()
```

```python
df['Message'] = df['Message'].str.lower()  #приводим к нижнему регистру
df['Message'] = df['Message'].replace('[\W_]+',' ', regex = True) # убираем все служебные символы
df['Message'] = df['Message'].apply(lambda x: [word for word in x.split() if word not in stopwords_set]) # убираем стоп-слова
df['Message'] = df['Message'].apply(lambda x: [wordnet_lemmatizer.lemmatize(word) for word in x]) #приводим слова к словарной форме
```
